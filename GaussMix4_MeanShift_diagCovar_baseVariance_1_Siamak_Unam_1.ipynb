**What this program is doing:**
Testing clustering replicability via bootstrap (BS) on **one (synthetic) dataset** of n samples.

Drawn from Gaussian mixture with 4 components.

An 80-20% split of the training-test data, via percent_train=0.8.

MeanShift used, have set k = 4.

B is the # of bootstrap (BS) iterations. First feed training data, then the cluster-identity decisions of the left-out (LO) data is recorded in the matrix "siamak_MeanShift_predicted_matrix" (B x # of test points) and the matrix is written to the file "MeanShift_predicted_testdata_80_20_1.csv."
Each BS consists of a different 80-20% split.
The output at each BS iteration will consist of the # of LO datapoints:  

1) their identities (i.e. values) are in the file "matrix_data_xxx.csv" with xxx the BS iteration #, and each of these files is of dimensionality : # of test points x dim.

2) the cluster # each datapoint has been assigned to  MeanShift_predicted_testdata_80_20_1.csv. It should be of dimensionality : B x # of test points.

3) the file samples_original_1.csv contains the original n datapoints (each is had dimensionality dim), so its dimensionality is: n x dim

4) Siamak_comment: the file clam.csv. Each column corresponds to the cluster ID (i.e. k columns). Each row corresponds to the identify of the datapoint in the original dataset (i.e. n rows). The entries of this file are the # of times that datapoint has been assigned to the specific cluster. The max # of any entrie would be B.

```python
import random
import csv
import shutil

import numpy as np
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import math

from sklearn.cluster import MeanShift, estimate_bandwidth
%matplotlib inline

seed = 123
#seed = 124

np.random.seed(seed)
random.seed(seed)

dim = 400
#dim = 200
#dim = 100
#dim = 10
#dim = 3

n = 10000
#n = 1000
#n = 500
#n = 100

#B = 2000  # number of BS iterations
#B = 1000  # number of BS iterations
#B = 300  # number of BS iterations
#B = 200  # number of BS iterations
#B = 100  # number of BS iterations
#B = 50  # number of BS iterations
#B = 20  # number of BS iterations
B = 10  # number of BS iterations
#B = 5  # number of BS iterations
#B = 1

#n_clusters_1 = 4        # number of clusters assumed

percent_train = 0.8
train_num_1 = int(n * percent_train)
test_num_1 = n - train_num_1

siamak_MeanShift_predicted_matrix = np.zeros((B,test_num_1))

NumberOfMixtures = 4

mean_vector_1 = np.ones(dim)
mean_vector_2 = 4 * np.ones(dim)
mean_vector_3 = 7 * np.ones(dim)
mean_vector_4 = 10 * np.ones(dim)

c1 = 1     #scale factor for mean
#c = 0.01

c2 = 1     #scale factor for variance
#c2 = 0.1   #low variance case
#c2 = 10     #high variance case

covariance_diag_1 = c1 * np.identity(dim)
covariance_diag_2 = c1 * np.identity(dim)
covariance_diag_3 = c1 * np.identity(dim)
covariance_diag_4 = c1 * np.identity(dim)

# Mixture weights (non-negative, sum to 1)
w = [0.25, 0.25, 0.25, 0.25]

# Mean vectors and covariance matrices
MeanVectors = [mean_vector_1, mean_vector_2, mean_vector_3, mean_vector_4]
CovarianceMatrices = [c2 * covariance_diag_1, c2 * covariance_diag_2, c2 * covariance_diag_3 , c2 * covariance_diag_4]

# Initialize arrays
samples = np.empty( (n,dim) ); samples[:] = np.NaN
componentlist = np.empty( (n,1) ); componentlist[:] = np.NaN

# Generate samples
for iter in range(n):
    # Get random number to select the mixture component with probability according to mixture weights
    DrawComponent = random.choices(range(NumberOfMixtures), weights=w, cum_weights=None, k=1)[0]
    # Draw sample from selected mixture component
    DrawSample = np.random.multivariate_normal(MeanVectors[DrawComponent], CovarianceMatrices[DrawComponent], 1)
    # Store results
    componentlist[iter] = DrawComponent
    samples[iter, :] = DrawSample

# Report fractions
print('Fraction of mixture component 0:', np.sum(componentlist==0)/n)
print('Fraction of mixture component 1:',np.sum(componentlist==1)/n)
print('Fraction of mixture component 2:',np.sum(componentlist==2)/n)
print('Fraction of mixture component 3:',np.sum(componentlist==3)/n)

# Visualize result
plt.plot(samples[:, 0], samples[:, 1], '.', alpha=0.5)
plt.grid()
plt.show()

# Visualize result - Siamak added 3/7/24
plt.plot(samples[:, 0], samples[:, (dim-1)], '.', alpha=0.5)
plt.grid()
plt.show()
```
